```python
dataset = dataset.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])
dataset.head()
```

```python
dataset['v1'] = dataset['v1'].apply(lambda x: 0 if x == 'ham' else 1)
dataset.head()
```

```python
import en_core_web_md
nlp = en_core_web_md.load()
```

```python
# Import Stop words 
from spacy.lang.en.stop_words import STOP_WORDS
```

```python

dataset["good_msg"] = dataset["v2"].apply(lambda x:''.join(ch for ch in x if ch.isalnum() or ch==" " or ch=="'"))
dataset["good_msg"] = dataset["good_msg"].apply(lambda x: x.replace(" +"," ").lower().strip())
dataset["good_msg"] = dataset["good_msg"].apply(lambda x: " ".join([token.lemma_ for token in nlp(x) if (token.lemma_ not in STOP_WORDS) and (token.text not in STOP_WORDS)]))
dataset.head()
```

```python
mask = dataset.good_msg.apply(lambda x: type(x)==str)
mask.value_counts()
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=1000, oov_token="out_of_vocab") # instanciate the tokenizer
tokenizer.fit_on_texts(dataset.good_msg)
dataset["msg_encoded"] = tokenizer.texts_to_sequences(dataset.good_msg)
```

```python
dataset.head()
```

```python
reviews_pad = tf.keras.preprocessing.sequence.pad_sequences(dataset.msg_encoded, padding="post")
```

```python
# Train Test Split
xtrain, xval, ytrain, yval = train_test_split(reviews_pad,dataset.v1, test_size=0.3)
```

```python
train_batch = train.shuffle(len(train)).batch(64)
val_batch = val.shuffle(len(val)).batch(64)
```
